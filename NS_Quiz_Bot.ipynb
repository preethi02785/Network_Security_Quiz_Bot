```python
!pip install -q langchain openai tiktoken PyPDF2 faiss-cpu

!pip install -q langchain openai tiktoken PyPDF2 faiss-cpu

import os
os.environ["OPENAI_API_KEY"] = "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"

from PyPDF2 import PdfReader
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.text_splitter import CharacterTextSplitter
from langchain.vectorstores import FAISS
from langchain.chains.question_answering import load_qa_chain
from langchain.llms import OpenAI
import random

# Function to generate questions from text
def generate_questions(text):
    # You can use your preferred method to generate questions here
    # For simplicity, let's split the text into sentences and use them as questions
    sentences = text.split('.')
    questions = [f"What is {sentence}?" for sentence in sentences if sentence.strip()]
    return questions

doc_reader = PdfReader(r"/content/NS_QUIZBOT_data.pdf")

extracted_text = ''
for i, page in enumerate(doc_reader.pages):
    text = page.extract_text()
    if text:
        extracted_text += text

text_splitter = CharacterTextSplitter(
    separator="\n",
    chunk_size=650,
    chunk_overlap=100,
    length_function=len,
)
texts = text_splitter.split_text(extracted_text)

embeddings = OpenAIEmbeddings()
docsearch = FAISS.from_texts(texts, embeddings)

# Generate questions from the extracted text
questions = generate_questions(extracted_text)

# Load the question-answering chain

chain = load_qa_chain(OpenAI(), chain_type="stuff")

while True:
    # Randomly select a question
    random_question = random.choice(questions)

    # Print the random question and ask the user for an answer
    print(f"Random Question: {random_question}")
    user_input = input("Your answer (type 'stop' to exit or 'next' for the next question): ")

    if user_input.lower() == "stop":
        print("Stopping the code as per your request.")
        break
    elif user_input.lower() == "next":
        continue

    # Provide the user's answer as input to the chain
    docs = docsearch.similarity_search(random_question)
    response = chain.run(input_documents=docs, question=random_question, user_answer=user_input)
    print("Answer:",response)
```
